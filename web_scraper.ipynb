```python
# 3 Modules we are going to use are as follows 
# BeautifulSoup
# lxml parser  -> used to parse data to html. Like from xml, json etc to html
# requests -> used to request data from a website
```

```python
from bs4 import BeautifulSoup # This allows us to scrap a website. Gets the data and downloads the webpage.
import requests
import pandas as pd
from pandas import Series, DataFrame
```

```python
url = 'https://www.ndtv.com/coronavirus'
```

```python
# 1. Call The API using requests module
result = requests.get(url)
# print(result) # this is the output -> <Response [200]>
# print(result.content) # output is a long string which is messy like -> Data science - Wikipedia</title>\n<script>document.documentElement.className="client-js";RLCONF={"wgBreakFrames":!1,

soup = BeautifulSoup(result.content, 'lxml')  # lxml is the parser here
```

```python
print(soup) # We will get the clean html
```

```python
content = soup.find('div') # instead of div, we can find section too.
```

```python
print(content)   # we can do content[0] then content[1] etc too. to find length of number of divs we can do -> len(content)
```

```python
summary = content.find(class_ = 'myth-wrap')  # there is a findAll method too
```

```python
print(summary)
```

```python
print(summary.findAll('h2'))
```

```python
l = summary.findAll('h2')
```

```python
series = [ele.get_text() for ele in l]   
```
